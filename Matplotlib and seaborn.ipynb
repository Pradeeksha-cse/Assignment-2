```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
plt.hist(df["Glucose"])
plt.title("Distribution of Glucose Levels")
plt.xlabel("Glucose")
plt.ylabel("Frequency")
plt.show()
```

```python
df["Outcome"].value_counts().plot(kind="bar")
plt.title("Diabetes Outcome Count")
plt.xlabel("Outcome (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
plt.plot(df["Glucose"][:50])
plt.title("Glucose Trend (First 50 Records)")
plt.xlabel("Index")
plt.ylabel("Glucose")
plt.show()
```

```python
plt.scatter(df["BMI"], df["Glucose"])
plt.title("BMI vs Glucose")
plt.xlabel("BMI")
plt.ylabel("Glucose")
plt.show()
```

```python
plt.fill_between(df.index[:50], df["Glucose"][:50])
plt.title("Glucose Area Plot")
plt.xlabel("Index")
plt.ylabel("Glucose")
plt.show()
```

```python
plt.pie(df["Outcome"].value_counts(), labels=["No Diabetes","Diabetes"], autopct="%1.1f%%")
plt.title("Diabetes Distribution")
plt.show()
```

```python
sns.countplot(x="Outcome", data=df)

plt.title("Diabetes Count")
plt.xlabel("Outcome (0 = No, 1 = Yes)")
plt.ylabel("Count")

plt.show()
```

```python
sns.boxplot(x="Outcome", y="Glucose", data=df)
plt.show()
```

```python
sns.histplot(df["Glucose"])
plt.show()
```

```python
df["Preg_Category"] = pd.cut(df["Pregnancies"],
                            bins=[0,2,5,10,20],
                            labels=["Low","Medium","High","Very High"])
counts = df["Preg_Category"].value_counts()
plt.pie(counts, labels=counts.index, autopct="%1.1f%%")
plt.title("Pregnancies Distribution")
plt.show()
```
